```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head(10)
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=20)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.pairplot(tips,hue='sex',palette='gnuplot')
```

```python
import numpy as np
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,color='blue',hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,color='black',jitter=True,hue='sex',dodge=True)
```

```python
sns.factorplot(x='day',y='total_bills',data=tips,k)
```
