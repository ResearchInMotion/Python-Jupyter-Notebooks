```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
axes.set_ylabel('y')
axes.set_title('title');
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="black", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B" , alpha=1)        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")
```

```python
fig, ax = plt.subplots(figsize=(12,6))


# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=1, ls=':')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='Red',lw=5,linestyle=':',marker='s',markersize=10)
```
